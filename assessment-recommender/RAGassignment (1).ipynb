```python
def clean_test_type(tt):
    print("Raw input:", repr(tt))
    if isinstance(tt, str):
        cleaned = [x.strip() for x in tt.split(',') if x.strip()]
        print("Cleaned output:", cleaned)
        return cleaned
    return []

pre['Test Type Clean'] = pre['Test Type'].apply(clean_test_type)
```

```python
def row_to_text(row):
    return f"""
    Solution Name: {row['Solution Name']}
    Source: {row['Source']}
    Remote Testing: {row['Remote Testing']}
    Adaptive / IRT: {row['Adaptive / IRT']}
    Test Types: {', '.join(row['Test Type Description'])}
    """
```

```python
faiss_index = faiss.IndexFlatL2(doc_embeddings.shape[1])
faiss_index.add(doc_embeddings)
```

```python
model_name = "google/flan-t5-base"  # much faster, lighter
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name).to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
def generate_response(query, top_docs):
    prompt = f"""You are an expert in designing assessment strategies for hiring in the banking sector.

Based on the following documents, suggest detailed assessments suitable for evaluating candidates for entry-level banking roles such as cashiers.

User Query:
{query}

Relevant Documents:
{top_docs}

Provide your answer in a clear bullet list format. Be specific and avoid vague suggestions.
"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=512,
            temperature=0.7,
            top_p=0.9,
            do_sample=True,
            eos_token_id=tokenizer.eos_token_id
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response.split("Be specific and avoid vague suggestions.")[-1].strip()
```
